 # Graph Plotting using Histogram through matplotlib.pyplot

```python

# Step 4: Extract specific columns for plotting
sepal = df['SepalLengthCm']
leaf = df['PetalLengthCm']
ws = df['SepalWidthCm']
wp = df['PetalWidthCm']
# Step 5: Plot histograms for the extracted columns
hist_plot = plt.hist([sepal, leaf, ws, wp], 
                     orientation='vertical', 
                     color=['b', 'g', 'y', 'm'], 
                     label=['SepalLengthCm', 'PetalLengthCm', 'SepalWidthCm', 'PetalWidthCm'], 
                     width=0.5)
# Add title and labels for the histogram plot
plt.title('Sepal Length Calculation')
plt.ylabel('Count')
plt.xlabel('Measurement')
plt.legend()
plt.show()
```

# Plotting scatter plot using matplotlib.pyplot 

```python
# Step 6: Plot scatter plots for the extracted columns
plt.scatter(sepal, leaf, label='Sepal vs Petal Length')
plt.scatter(sepal, ws, label='Sepal Length vs Sepal Width')
plt.scatter(sepal, wp, label='Sepal Length vs Petal Width')
# Add title and labels for the scatter plot
plt.title('Scatter Plot of Sepal and Petal Measurements')
plt.xlabel('Sepal Length (Cm)')
plt.ylabel('Measurements (Cm)')
plt.legend()
plt.show()
```
